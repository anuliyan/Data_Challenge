# How many Pageviews did Glamour have in August?

```python
#converting date into year/month/day
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
#filitering to month of auguest
August_df = df.loc[(df['Date'] >= '2019-07-30')
                     & (df['Date'] < '2019-09-01')]
```

```python
#Finding pitchfork on auguest
Pitchfork_brand = August_df.loc[August_df['Brand'] == 'Glamour']
```

```python
#Finding eventname Pageview
Glamour_brand_pageview = Pitchfork_brand.loc[Pitchfork_brand['Event_Name'] == 'Pageview']
```

# #3 What percentage of SELF’s sessions are bounces. Which device type has the highest
bounce rate? We define a “bounce” as a session with only one pageview.


```python
selff = df.loc[df['Brand'] == 'Self']
```

```python
pageview = selff.loc[selff['Event_Name'] == 'Pageview']
```

```python
pageview_Event_order = pageview.loc[pageview['Event_Order'] == 1]
```

```python
# Desktops has the highest bounce rates
pageview_Event_order['Device'].value_counts()
```

```python
selff['Event_Order'].value_counts()
```

```python
# Calculating Percentage
selff["percent"] = (selff['Event_Order'] / selff['Event_Order'].sum()) * 100
```

```python
# Bounce Rate is 0.000952
selff["percent"].round(100)
```
