```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#opening the file
df = pd.read_excel(r'/Users/anuliyanagamage/Downloads/Data_Challenge_SRPA.xlsx') 
```

```python
df
```

```python
# Creating a dataset with the CNT Brand
CNT_brand = df.loc[df['Brand'] == 'Conde Nast Traveler']
```

```python
CNT_brand
```

```python
#filitering to eventname to click
CNT_brand = CNT_brand.loc[(CNT_brand['Component_Name'] == 'Advertisement') & (CNT_brand['Event_Name'] == 'Click')]
```

```python
# Food has been the most frequent in site section
CNT_brand['Site_Section'].describe()
```

```python
# Most frquent used device is a desktop
CNT_brand['Device'].describe()
```

```python
# India has is the top Market for clicks
CNT_brand['Market'].describe()
```

```python
CNT_brand
```

```python
# Filitering month 
CNT_brand['month'] = CNT_brand['Date'].dt.month
```

```python
sessions = CNT_brand.groupby(['month'])['Session_Id'].nunique()
```

```python
# Following graph shows the rate of clicked adventisments per month
```

```python
# Sessions Per Month
sessions.plot(x="Month", y=["Session_Id"], kind="line",label='Sessions')
plt.legend()
plt.show()
```

* Desktop users are more likly to click on CNT ads compared to other devices.
* Markets such as India will have frequent users with ad clicks. 
* Site selections of foods are great way to target for CNT ad clicks.
